**Objetivo da Atividade**

Exercitar a lógica de programação por meio de pseudocódigo para resolver um problema real e atual do cotidiano brasileiro: o monitoramento dos preços da cesta básica. O desafio consiste em aplicar estruturas de decisão, manipulação de variáveis e entradas simples para simular uma ferramenta de análise de variação de preços.

**O que seu algoritmo deve fazer:**

- **Ler:** nome do produto, preço anterior e preço atual.
- **Calcular:** a variação percentual do preço.
- **Exibir:**  
    - Se o preço aumentou, diminuiu ou permaneceu estável;  
    - Se o aumento foi superior a 10%, sinalizar como abuso.

> Na primeira parte do exercício, utilizei a biblioteca `random` para gerar preços aleatórios dentro de intervalos pré-determinados para cada produto. Armazenei esses valores em listas e empreguei a estrutura de repetição `for` para simular os preços ao longo de 12 meses.

```python
import random
import datetime

lista_arroz = [] # LISTA DE PREÇOS ARROZ
lista_feijao = [] # LISTA DE PREÇOS FEIJÃO
lista_acucar = [] # LISTA DE PREÇOS AÇÚCAR


for mes in range(1,13): # GERAR DADOS PARA 12 MESES
   valores_arroz = random.uniform(19.0, 30.0) # GERAR VALORES ALEATÓRIOS ENTRE 19 E 30 REAIS PARA ARROZ
   lista_arroz.append([mes, valores_arroz]) # ADICIONAR MÊS E VALOR NA LISTA ARROZ

   valores_feijao = random.uniform(5.0, 10.0) # GERAR VALORES ALEATÓRIOS ENTRE 5 E 10 REAIS PARA FEIJÃO
   lista_feijao.append([mes, valores_feijao]) # ADICIONAR MÊS E VALOR NA LISTA FEIJÃO

   valores_acucar = random.uniform(3.0, 8.0) # GERAR VALORES ALEATÓRIOS ENTRE 3 E 8 REAIS PARA AÇÚCAR
   lista_acucar.append([mes, valores_acucar]) # ADICIONAR MÊS E VALOR NA LISTA AÇÚCAR

print(lista_arroz) # IMPRIMIR LISTA DE PREÇOS ARROZ
print(lista_feijao) # IMPRIMIR LISTA DE PREÇOS FEIJÃO
print(lista_acucar) # IMPRIMIR LISTA DE PREÇOS AÇÚCAR
```

Na segunda parte do exercício, utilizei a estrutura de repetição for para percorrer as listas de preços ao longo dos meses. Empreguei estruturas condicionais para comparar os preços de cada mês com o mês anterior, identificando aumentos, reduções ou estabilidade nos valores. Para facilitar a visualização, implementei um sistema de cores: 
- **vermelho para aumentos abusivos acima de 10%**
- **azul para aumentos abaixo de 10%**
- **verde para reduções de preço**
- **cor neutra caso não haja alteração**

```python
for c in range(0 ,len(lista_arroz)): # PERCORRER A LISTA DE 12 MESES
    if c != 0: # IGNORAR O PRIMEIRO MÊS POIS NÃO TEM MÊS ANTERIOR PARA COMPARAR
      if lista_arroz[c][1] > (lista_arroz[c-1][1]*1.1): # VERIFICAR SE O PREÇO DO MÊS ATUAL É MAIOR QUE 10% DO MÊS ANTERIOR
         print(f"\033[91mNo mês {lista_arroz[c][0]} o Arroz teve aumento de {((lista_arroz[c][1] - lista_arroz[c-1][1]) / lista_arroz[c-1][1]) * 100:.2f}%, aumento acima de 10% em relação ao mês {lista_arroz[c-1][0]}.\033[0m") # IMPRIMIR MENSAGEM DE AUMENTO ABUSIVO
         print(f"Preço mês atual R${lista_arroz[c][1]:.2f}") # IMPRIMIR PREÇO ATUAL
         print(f"Preço mês anterior R${lista_arroz[c-1][1]:.2f}") # IMPRIMIR PREÇO ANTERIOR
         print(f"\033[91mAUMENTO ABUSIVO ACIMA DE 10%\033[0m") # IMPRIMIR ALERTA DE AUMENTO ABUSIVO
         print("")

      elif lista_arroz[c][1] < lista_arroz[c-1][1]: # VERIFICAR SE O PREÇO DO MÊS ATUAL É MENOR QUE O MÊS ANTERIOR
         print(f"\033[92mNo mês {lista_arroz[c][0]} o Arroz teve redução de {((lista_arroz[c-1][1] - lista_arroz[c][1]) / lista_arroz[c-1][1]) * 100:.2f}% em relação ao mês {lista_arroz[c-1][0]}.\033[0m") # IMPRIMIR MENSAGEM DE REDUÇÃO DE PREÇO
         print(f"Preço mês atual R${lista_arroz[c][1]:.2f}") # IMPRIMIR PREÇO ATUAL
         print(f"Preço mês anterior R${lista_arroz[c-1][1]:.2f}") # IMPRIMIR PREÇO ANTERIOR
         print(f"\033[92mTEVE REDUÇÃO DE PREÇO\033[0m") # IMPRIMIR ALERTA DE REDUÇÃO DE PREÇO
         print("")

      elif lista_arroz[c][1] < (lista_arroz[c-1][1]*1.1) and lista_arroz[c][1] > lista_arroz[c-1][1]: # VERIFICAR SE O PREÇO DO MÊS ATUAL É MAIOR QUE O MÊS ANTERIOR MAS MENOR QUE 10%
         print(f"\033[94mNo mês {lista_arroz[c][0]} o Arroz teve aumento de {((lista_arroz[c][1] - lista_arroz[c-1][1]) / lista_arroz[c-1][1]) * 100:.2f}%, aumento abaixo de 10% em relação ao mês {lista_arroz[c-1][0]}.\033[0m") # IMPRIMIR MENSAGEM DE AUMENTO DENTRO DO PERMITIDO 
         print(f"Preço mês atual R${lista_arroz[c][1]:.2f}") # IMPRIMIR PREÇO ATUAL
         print(f"Preço mês anterior R${lista_arroz[c-1][1]:.2f}") # IMPRIMIR PREÇO ANTERIOR
         print(f"\033[94mAUMENTO DENTRO DO PERMITIDO ABAIXO DE 10%\033[0m") # IMPRIMIR ALERTA DE AUMENTO DENTRO DO PERMITIDO
         print("")

      else: # CASO NÃO TENHA TEVE NENHUMA ALTERAÇÃO DE PREÇO
         print(f"No mês {lista_arroz[c][0]} o Arroz não teve alteração de preço em relação ao mês {lista_arroz[c-1][0]}.") # IMPRIMIR MENSAGEM DE NENHUMA ALTERAÇÃO
         print(f"Preço mês atual R${lista_arroz[c][1]:.2f}") # IMPRIMIR PREÇO ATUAL
         print(f"Preço mês anterior R${lista_arroz[c-1][1]:.2f}") # IMPRIMIR PREÇO ANTERIOR
         print("")



      if lista_feijao[c][1] > (lista_feijao[c-1][1]*1.1): # VERIFICAR SE O PREÇO DO MÊS ATUAL É MAIOR QUE 10% DO MÊS ANTERIOR
         print(f"\033[91mNo mês {lista_feijao[c][0]} o Feijão teve aumento de {((lista_feijao[c][1] - lista_feijao[c-1][1]) / lista_feijao[c-1][1]) * 100:.2f}%, aumento acima de 10% em relação ao mês {lista_feijao[c-1][0]}.\033[0m") # IMPRIMIR MENSAGEM DE AUMENTO ABUSIVO
         print(f"Preço mês atual R${lista_feijao[c][1]:.2f}") # IMPRIMIR PREÇO ATUAL
         print(f"Preço mês anterior R${lista_feijao[c-1][1]:.2f}") # IMPRIMIR PREÇO ANTERIOR
         print(f"\033[91mAUMENTO ABUSIVO ACIMA DE 10%\033[0m") # IMPRIMIR ALERTA DE AUMENTO ABUSIVO
         print("")

      elif lista_feijao[c][1] < lista_feijao[c-1][1]: # VERIFICAR SE O PREÇO DO MÊS ATUAL É MENOR QUE O MÊS ANTERIOR
         print(f"\033[92mNo mês {lista_feijao[c][0]} o Feijão teve redução de {((lista_feijao[c-1][1] - lista_feijao[c][1]) / lista_feijao[c-1][1]) * 100:.2f}% em relação ao mês {lista_feijao[c-1][0]}.\033[0m") # IMPRIMIR MENSAGEM DE REDUÇÃO DE PREÇO
         print(f"Preço mês atual R${lista_feijao[c][1]:.2f}") # IMPRIMIR PREÇO ATUAL
         print(f"Preço mês anterior R${lista_feijao[c-1][1]:.2f}") # IMPRIMIR PREÇO ANTERIOR
         print("")

      elif lista_feijao[c][1] < (lista_feijao[c-1][1]*1.1) and lista_feijao[c][1] > lista_feijao[c-1][1]: # VERIFICAR SE O PREÇO DO MÊS ATUAL É MAIOR QUE O MÊS ANTERIOR MAS MENOR QUE 10%
         print(f"\033[94mNo mês {lista_feijao[c][0]} o Feijão teve aumento de {((lista_feijao[c][1] - lista_feijao[c-1][1]) / lista_feijao[c-1][1]) * 100:.2f}%, aumento abaixo de 10% em relação ao mês {lista_feijao[c-1][0]}.\033[0m") # IMPRIMIR MENSAGEM DE AUMENTO DENTRO DO PERMITIDO
         print(f"Preço mês atual R${lista_feijao[c][1]:.2f}") # IMPRIMIR PREÇO ATUAL
         print(f"Preço mês anterior R${lista_feijao[c-1][1]:.2f}") # IMPRIMIR PREÇO ANTERIOR
         print("")   

      else: # CASO NÃO TENHA TEVE NENHUMA ALTERAÇÃO DE PREÇO
         print(f"No mês {lista_feijao[c][0]} o Feijão não teve alteração de preço em relação ao mês {lista_feijao[c-1][0]}.") # IMPRIMIR MENSAGEM DE NENHUMA ALTERAÇÃO
         print(f"Preço mês atual R${lista_feijao[c][1]:.2f}") # IMPRIMIR PREÇO ATUAL
         print(f"Preço mês anterior R${lista_feijao[c-1][1]:.2f}") # IMPRIMIR PREÇO ANTERIOR
         print("")


      if lista_acucar[c][1] > (lista_acucar[c-1][1]*1.1): # VERIFICAR SE O PREÇO DO MÊS ATUAL É MAIOR QUE 10% DO MÊS ANTERIOR
         print(f"\033[91mNo mês {lista_acucar[c][0]} o Açúcar teve aumento de {((lista_acucar[c][1] - lista_acucar[c-1][1]) / lista_acucar[c-1][1]) * 100:.2f}%, aumento acima de 10% em relação ao mês {lista_acucar[c-1][0]}.\033[0m") # IMPRIMIR MENSAGEM DE AUMENTO ABUSIVO
         print(f"Preço mês atual R${lista_acucar[c][1]:.2f}") # IMPRIMIR PREÇO ATUAL
         print(f"Preço mês anterior R${lista_acucar[c-1][1]:.2f}") # IMPRIMIR PREÇO ANTERIOR
         print(f"\033[91mAUMENTO ABUSIVO ACIMA DE 10%\033[0m") # IMPRIMIR ALERTA DE AUMENTO ABUSIVO
         print("")

      elif lista_acucar[c][1] < lista_acucar[c-1][1]: # VERIFICAR SE O PREÇO DO MÊS ATUAL É MENOR QUE O MÊS ANTERIOR
         print(f"\033[92mNo mês {lista_acucar[c][0]} o Açúcar teve redução de {((lista_acucar[c-1][1] - lista_acucar[c][1]) / lista_acucar[c-1][1]) * 100:.2f}% em relação ao mês {lista_acucar[c-1][0]}.\033[0m")
         print(f"Preço mês atual R${lista_acucar[c][1]:.2f}") # IMPRIMIR PREÇO ATUAL
         print(f"Preço mês anterior R${lista_acucar[c-1][1]:.2f}") # IMPRIMIR PREÇO ANTERIOR
         print("")

      elif lista_acucar[c][1] < (lista_acucar[c-1][1]*1.1) and lista_acucar[c][1] > lista_acucar[c-1][1]: # VERIFICAR SE O PREÇO DO MÊS ATUAL É MAIOR QUE O MÊS ANTERIOR MAS MENOR QUE 10%
         print(f"\033[94mNo mês {lista_acucar[c][0]} o Açúcar teve aumento de {((lista_acucar[c][1] - lista_acucar[c-1][1]) / lista_acucar[c-1][1]) * 100:.2f}%, aumento abaixo de 10% em relação ao mês {lista_acucar[c-1][0]}.\033[0m") # IMPRIMIR MENSAGEM DE AUMENTO DENTRO DO PERMITIDO
         print(f"Preço no mês atual R${lista_acucar[c][1]:.2f}") # IMPRIMIR PREÇO ATUAL
         print(f"Preço mês anterior R${lista_acucar[c-1][1]:.2f}") # IMPRIMIR PREÇO ANTERIOR
         print("")

      else: # CASO NÃO TENHA TEVE NENHUMA ALTERAÇÃO DE PREÇO
         print(f"No mês {lista_acucar[c][0]} o Açúcar não teve alteração de preço em relação ao mês {lista_acucar[c-1][0]}.") # IMPRIMIR MENSAGEM DE NENHUMA ALTERAÇÃO
         print(f"Preço mês atual R${lista_acucar[c][1]:.2f}") # IMPRIMIR PREÇO ATUAL
         print(f"Preço mês anterior R${lista_acucar[c-1][1]:.2f}") # IMPRIMIR PREÇO ANTERIOR
         print("")   

      print(200*"-")
```
